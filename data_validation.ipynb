```python
import pandas as pd
import matplotlib.pyplot as plt

# Load processed data
train = pd.read_csv(r"D:\gcetts\new project\hinglish-sentiment-analysis\data\processed\train.csv")

# Check for missing values
print(f"Missing values in the dataset: \n{train.isnull().sum()}")

# Drop rows with missing values in the 'text' or 'emotion' column
train.dropna(subset=['text', 'emotion'], inplace=True)

# Check class distribution
plt.figure(figsize=(10,5))
train['emotion'].value_counts().plot(kind='bar')
plt.title('Class Distribution')
plt.show()

# Check Hinglish samples
hinglish_samples = train[train['text'].str.contains(r'[a-z]+ [a-z]+ [a-z]+')]
print(f"Hinglish samples: {len(hinglish_samples)}/{len(train)}")

# Sample review: Ensure we don't sample more than the available data
sample_size = min(5, len(train))
print(f"\nSample entries (sampling {sample_size} entries):")
print(train.sample(sample_size))
```
